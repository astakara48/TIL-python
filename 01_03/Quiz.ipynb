```python
# 1번
import requests
from bs4 import BeautifulSoup
import urllib.request as req
import re
url="https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start=1"
page = req.urlopen(url)
doc=page.read()
soup=BeautifulSoup(doc, "html.parser")
questions=soup.find(id="elThumbnailResultArea")
questions_list=questions.find_all("dt", class_="question")
questions_list=questions.find_all("a")
for question in questions_list:
    if "href" in question.attrs:
        print(question.text)
```

```python
# 2번 
import requests
from bs4 import BeautifulSoup
import urllib.request as req
import re
url="https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start=1"
page = req.urlopen(url)
soup=BeautifulSoup(page, "html.parser")
links=[]
cnt=0
for link in soup.select("ul.type01 li dl dd"):
    links.append(link.text)
#print(links)
for a in links:
    if (cnt-1)%4==0:
        print(a)
    cnt+=1
```
