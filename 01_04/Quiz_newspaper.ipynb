```python
# 국민일보 페이지 넘버링 - http://www.kmib.co.kr/search/searchResult.asp?searchWord=%uBD09%uC900%uD638&pageNo=1&period=
import requests
from bs4 import BeautifulSoup
import urllib.request as req
import urllib
import re

url="http://www.kmib.co.kr/search/searchResult.asp?searchWord=%uBD09%uC900%uD638&pageNo="
url_back_part="&period="
current_url=""
```

```python
def get_url_from_title():
    num_url = "http://www.kmib.co.kr/search/searchResult.asp?searchWord=%uBD09%uC900%uD638&pageNo=1&period="
    res=req.urlopen(num_url)
    soup=BeautifulSoup(res, "html.parser")
    p=soup.select_one("#searchTotalItemCnt").text
    num=int(int(p)/10+1)
    for i in range(num):
        current_page_num=1+int(i)
        current_url = url+str(current_page_num)+url_back_part
        # print(current_url)
        page=req.urlopen(current_url)
        doc=page.read()
        #print(doc)
        soup=BeautifulSoup(doc,"html.parser")
        #print(soup)
        titles=soup.find(id="searchList")
        #print(titles)
        title_list=titles.find_all("dt", class_="tit")
        for title in title_list:
            title_link = title.select("a")
            article_url = title_link[0]["href"]
            get_text(article_url)

            
def get_text(article_url):
    res=req.urlopen(article_url).read()
    content=res.decode('euc-kr','replace').encode("utf-8","replace")
    soup=BeautifulSoup(content,"html.parser")
    bodys=soup.find(id="article")
    body_list=bodys.find_all(class_="tx")
    for body in body_list:
        print(body.text)

get_url_from_title()
```
