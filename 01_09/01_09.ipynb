```python
df_left=pd.DataFrame({
    "KEY":["K0","K1","K2","K3"],
    "A":["A0","A1","A2","A3"],
    "B":[0.5,1.0,0.7,1.5],
})
df_right=pd.DataFrame({
    "KEY":["K2","K3","K4","K5"],
    "C":["C2","C3","C4","C5"],
    "D":["D2","D3","D4","D5"],
})
```

```python
# df_left, df_right를 머지, KEY를 기준으로, outer join
df_all=df_left.merge(df_right, on="KEY", how="outer")

```

```python
# DataFrame 전체에 대해 결측값 확인
# df.isnull(), isnull(df), notnull()
pd.isnull(df_all)
df_all.isnull()
pd.notnull(df_all)
df_all.notnull()
df_all
```

```python
df_all.ix[[0,1],["A","B"]]=None
df_all
```

```python
df_all.isnull().sum()
# zjffjaquf rufcmrrkqt rotn
# A열의 결측값 개수 출력
df_all["A"].isnull().sum()

# df_all 각 열별로 결측값이 아닌 개수 출력
df_all[:].notnull().sum() # default가 axis=0
df_all[:].notnull().sum(axis=1)
df_all
```

```python
df_all # 맨 우측에 NaN_cnt, NotNull_cnt 컬럼 추가
df_all["NaN_cnt"]=df_all.isnull().sum(1)
df_all
df_all["NotNull_cnt"]=df_all.notnull().sum(1)
df_all
```

```python
df.c1.sum() # sum함수 연산시 NaN은 0으로 처리
df["c1"].sum()
df.sum()
df["c1"].cumsum() # 아래로 내려가면서 누적으로 sum
```
