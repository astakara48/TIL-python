```python
# 년, 월, 일, 시에 따른 평균 대여량 구하기
train_2011=train[train["Year"].isin(["2011"])]
print("2011년도 평균 대여량: ",train_2011["count"].mean())

train_2012=train[train["Year"].isin(["2012"])]
print("2012년도 평균 대여량 : ",train_2012["count"].mean())

for i in range(1,13):
    train_m=train[train["Month"].isin([i])]
    print(i,"월 평균 대여량: ",train_m["count"].mean())

for i in range(1,20):
    train_d=train[train["Day"].isin([i])]
    print(i,"일 평균 대여량 : ",train_d["Day"].mean())
```
