```python
train['전년동월대비']=train['전년동월대비'].astype('str')
```

```python
# ▲▼ 같은 문자를 처리, ▲는 양수로, ▼는 음수로
up='▲'
down='▼'
comma=','
last_year_by_month=[]
for i in train['전년동월대비']:
    if comma in i:
        i=i.replace(",","") # 1,715▼ 같은 수치가 object라서 처리
        if up in i:
            loc=i.index(up)
            num=int(i[:loc])
            last_year_by_month.append(num)
        elif down in i:
            loc=i.index(down)
            num=int(i[:loc])
            last_year_by_month.append(-num)
    else:
        if up in i:
            loc=i.index(up)
            num=int(i[:loc])
            last_year_by_month.append(num)
        elif down in i:
            loc=i.index(down)
            num=int(i[:loc])
            last_year_by_month.append(-num)
        else:
            num=int(float(i))
            last_year_by_month.append(num)
```
