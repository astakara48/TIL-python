```python
# 자료형 처리
import pandas as pd
import seaborn as sns
import numpy as np
from pandas import DataFrame
import warnings
warnings.filterwarnings(action="ignore")
```

```python
unames=['user_id','gender','age','occupation','zip']
rnames=['user_id','movie_id','ratings','timestamp']
mnames=['movie_id','title','genres']

upath="dataset/movielens/users.dat"
rpath="dataset/movielens/ratings.dat"
mpath="dataset/movielens/movies.dat"

users=pd.read_csv(upath, sep="::", header=None, names=unames)
ratings=pd.read_csv(rpath, sep="::", header=None, names=rnames)
movies=pd.read_csv(mpath, sep="::", header=None, names=mnames)
```

```python
"""
ex)
x라는 사용자가 영화들을 보고 평점을 주었을때 비슷한 사람 찾기
영화ID  평점
3        4
5        1
9        0
10       5
15       4

1. 조건 3편 이상 중복 영화
2. 평점 유사하게 부여한 사람 => 유클리디안 기반의 거리 계산 방법 이용 : 루트[(x1-y1)^2+(x2-y2)^2...]
3. 비슷한 사람 z를 찾았다고 할때 // x가 안본 영화들중 z가 본 영화 목록들 중 평점을 높게 준 영화를 1번에게 추천 ==> CF 
"""
```

```python
data=pd.merge(pd.merge(ratings, users),movies)
data
```

```python
mean_ratings=pd.pivot_table(data, index='title', columns='gender', values='ratings', aggfunc=np.mean)
mean_ratings
```

```python
# title을 기준으로 그룹화
ratings_by_title=data.groupby('title').size()
type(ratings_by_title)
ratings_by_title.index
active_titles=ratings_by_title.index[ratings_by_title.values>=250]

```

```python
# mean_ratings에서 active_titles에 저장된 영화제목에 대한 데이터만 추출
mean_ratings.iloc[0]
mean_ratings=mean_ratings.loc[active_titles]
mean_ratings
```

```python
mean_ratings.sort_values(by='F', ascending=False) # F열을 기준으로 내림차순으로 정렬
```
