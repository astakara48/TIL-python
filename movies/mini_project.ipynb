```python
# 1번 신생아 이름 데이터를 이용하여 다음과 같은 분석을 진행하시오.
# -각 연도별 가장 인기 있는 이름을 분석
# -다음 이름의 연도별 순위 조사(john, sophia, michael, daniel, carter)
# -각 연도별, 성별 상위 50% 이름을 구성하는 알파벳 문자 중, 자음과 모음의 비율 분석
# -가장 긴 이름과 가장 짧은 이름을 출력
# -시각화도 함께 ...

pieces=[]
for year in range(1880, 2011):
    path="names/yob%d.txt" % year
    df=pd.read_csv(path, names=['name','sex','births'])
    df['year']=year
    pieces.append(df)
    
names=pd.concat(pieces, ignore_index=True)
names
```

```python
# 1-1 유명인, 성서에 등장하는 이름 등 외부 이름을 이용하여 이름 유행을 분석
s=['Jordan','Jackson','Madonna','Ronaldo']
for i in s:
    similar_celeb=names.loc[names.name==i]
    similar_celeb=pd.pivot_table(similar_celeb, index='year', columns='name', values='births')
    similar_celeb.plot()
```

```python
# 1-2 각 연도별 가장 인기 있는 이름을 분석
popular_names=[]
for i in range(1880, 2011):
    name_years=names.loc[names.year==i]
    popular_names.append(name_years.loc[name_years.births.rank(ascending=False)==1.0])

popular_names_year=pd.concat(popular_names)
popular_names_year
```

```python
# 1-3 다음 이름의 연도별 순위 조사(john, sophia, michael, daniel, carter)
s=['John','Sophia','Michael','Daniel','Carter']

pieces=[]
for year in range(1880, 2011):
    path="names/yob%d.txt" % year
    df=pd.read_csv(path, names=['name','sex','births'])
    df['year']=year
    df['rank']=df.births.rank(method='max', ascending=False)
    pieces.append(df)
    
names=pd.concat(pieces, ignore_index=True)

names_list=[]
for i in s:
    names_list.append(names.loc[names.name==i])
    
names_year=pd.concat(names_list)
names_year
names_year=pd.pivot_table(names_year, index='year', columns='name', values='rank')
names_year.plot(figsize=(16,8))
```

```python
# 1-4 각 연도별, 성별 상위 50% 이름을 구성하는 알파벳 문자 중, 자음과 모음의 비율 분석

boys=[]
girls=[]

for year in range(1880, 2011):
    path="names/yob%d.txt" % year
    df=pd.read_csv(path, names=['name','sex','births'])
    df['year']=year
    
    boy=df.loc[df.sex=='M']
    births=boy.births.sum()
    boy['prop']=boy.births/births
    
    girl=df.loc[df.sex=='F']
    births=girl.births.sum()
    girl['prop']=girl.births/births
    
    boy.sort_index(by='births', ascending=False)
    girl.sort_index(by='births', ascending=False)
    
    boy=boy[boy.sort_index(by='prop', ascending=False).prop.cumsum()<=0.5]
    
    girl=girl[girl.sort_index(by='prop', ascending=False).prop.cumsum()<=0.5]
    
    boys.append(boy)
    girls.append(girl)
    
boys=pd.concat(boys)
girls=pd.concat(girls)
```
