```python
# 이항변수화 : 연속형변수 기준이하 -> 0, 기준초과 -> 1
# 확률변수 x가 이항분포를 따른다고 했을때, 0 또는 1값을 갖는 이항변수화가 필요
# 베르누이시행 : 성공(1), 실패(0) 두가지 경우만 존재
# 성공확률이 p인 베르누이시행을 n번 수행 
# 성공하는 횟수를 x라고 하면, 확률변수 x는 모수 n과 p인 이항분포를 따른다.

"""
주사위 : 3
3이 나올 확률(성공확률) : 1/6
실패 : 5/6
A, B, C, D 
(A만 성공, B만 성공, C만 성공, D만 성공)
A만 성공 : 1/6*5/6*5/6*5/6=0.09
B만 성공 : 1/6*5/6*5/6*5/6=0.09
C만 성공 : 1/6*5/6*5/6*5/6=0.09
D만 성공 : 1/6*5/6*5/6*5/6=0.09
0.09*4 = 0.36
성공확률(p)*시나리오수(n)=이항분포

"""
```

```python
df = DataFrame({'C1': np.random.randn(20),

   'C2': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a',

   'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b']})
df
```

```python
"""
np.digitize함수, np.where함수 : 연속형 변수 => 이산화
이항변수화 : 0 또는 1 값을 갖는 변수를 만드는 것
이산화 : 연속형 변수를 2개 이상의 변주를 갖는 변수로 변환
ex) 점수 : 75->A/B/C/D/F
"""

np.random.seed(10) # 42, 777, 1004
np.random.randn(1)
# 동일한 seed 값을 주면 난수도 동일하게 나온다.
# seed값을 주고 random을 하는 이유는 일정한 값의 데이터를 얻기 위해
```

```python
"""
df의 c1열 20개 data 구성
10개 bin으로 균등하게 나눔
c1의 최대, 최소를 찾아서 20개로 균등하게 나눔
"""
df
```

```python
df.C1.min() # -2.5525293341518744
df.C1.max() # 1.820030089427802
bins=np.linspace(df.C1.min(), df.C1.max(),10) # 주어진 데이터의 범위를 균등하게 여러 구간으로 난루 수 있다. 시작값, 끝값, 범위
bins
df["C1_bins"]=np.digitize(df.C1, bins)
df.C1_bins
```

```python
# 구간별로 요약통계, 범주간 평균 차이
# 독립성검정, indexing 부여
df.groupby(df.C1_bins)['C1'].size() # 같은 범위끼리 그룹화 + size붙여서 그룹별로 몇개씩 있는지 확인
df.groupby(df.C1_bins)['C1'].mean() # 각 그룹 별로 평균값이 몇인지?
df.groupby(df.C1_bins)['C1'].std() # 데이터가 1건밖에 없으면 NaN이 나옴
df.groupby(df.C1_bins)['C2'].value_counts() # C1_bins를 기준으로 그룹화 하고, c2를 참고하고 

df[df['C1_bins']==4] # C1_bins가 4인 값만 따로 추출
```

```python
# pd.get_dummies() : 가변수 생성
pd.get_dummies(df["C1_bins"], prefix="C1") # 원핫인코딩 // prefix-> 접두어
# 모델=x1*가중치+x2*가중치+x10*가중치+바이어스
```

```python
# np.where(조건, 참, 거짓)
df["h_l"]=np.where(df["C1"]>=df.C1.mean(),"high","low")
df
df.groupby("h_l")["C1"].size()
df.groupby("h_l")["C1"].mean()
df.groupby("h_l")["C1"].std()
```

```python
Q1=np.percentile(df["C1"],25) # 1사분위의 수 추출 25%
Q3=np.percentile(df["C1"],75) # 3사분위의 수 추출 75%
# 연습문제 : C1열 값 <= Q1 then l
# Q1<C1열 값 <= Q3 then m
# else h
df["h_m_l"]=np.where(df["C1"]>=Q3,"high",np.where(df["C1"]<=Q1,"low","medium"))
df
```

```python
# 실수 데이터 => 카테고리화 
# cut(실수 값의 경계를 지정)
# qcut(개수가 동일한 똑같은 구간으로 나눌때)
ages=[0,2,10,21,23,37,31,61,20,42,32,100]
bins=[1,15,25,35,60,99]
labels=["미성년자","청년","중년","장년","노년"]
cats=pd.cut(ages, bins, labels=labels) # 데이터, 나눌 경계, 라벨
cats
cats.categories
cats.codes # 몇번쨰 범주에 해당되는지 index 번호
```

```python
df=pd.DataFrame(ages, columns=["ages"])
df["age_cat"]=pd.cut(ages,bins,labels=labels)
df
```
