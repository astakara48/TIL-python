```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import warnings
warnings.filterwarnings(action="ignore")
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import OneHotEncoder
```

```python
# 1번

# 1) Pclass를 원핫인코딩
taitanic=pd.read_csv("train.csv")
taitanic
taitanic.Pclass.max() # 3
taitanic.Pclass.min() # 1
bins=np.linspace(taitanic.Pclass.min(), taitanic.Pclass.max(), 3)
bins
taitanic["Pclass_bins"]=np.digitize(taitanic.Pclass, bins)
taitanic.Pclass_bins
pd.get_dummies(taitanic["Pclass_bins"], prefix="Pclass")
```

```python
# 2) Age를 이산화(5개 구간으로 나눔)
ages=taitanic.Age
ages=np.array(ages)
bins=[1,15,25,35,45,99]
labels=["미성년자","청년","중년","장년","노년"]
cats=pd.cut(ages, bins, labels=labels)
cats.categories
cats.codes
test_age=pd.DataFrame(ages, columns=["Age"])
test_age["age_cat"]=pd.cut(ages,bins,labels=labels)
test_age
```

```python
# 3) Fare를 이산화(3개 구간으로 나눔)
fare=taitanic.Fare
fare=np.array(fare)
np.max(fare) # 512.3292
np.min(fare) # 0.0
# 341.5468, 170.7734
bins=[0,10,25,512.3292]
labels=["3등석","2등석","1등석"]
cats=pd.cut(fare,bins,labels=labels)
cats.categories
cats.codes
cats
test_fare=pd.DataFrame(fare, columns=["Fare"])
test_fare["Fare_cat"]=pd.cut(fare,bins,labels=labels)
test_fare
```

```python
# 4번

ages=taitanic.Age
ages=np.array(ages)
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
survive=np.where(taitanic.Survived==1,taitanic.Age,np.nan)
death=np.where(taitanic.Survived==0,taitanic.Age,np.nan)
survive_list=pd.cut(survive, bins, labels)
death_list=pd.cut(death, bins, labels)

survive_list.categories
survive_list.codes
survive_ratio=pd.value_counts(survive_list)/pd.value_counts(survive_list).sum()

death_list.categories
death_list.codes
death_ratio=pd.value_counts(death_list)/pd.value_counts(death_list).sum()

print("생존자 비율")
print(survive_ratio)
print("="*50)
print("사망자 비율")
print(death_ratio)
print("="*50)
print("생존자 비율 총합 : ", survive_ratio.sum())
print("사망자 비율 총합 : ", death_ratio.sum())
```
