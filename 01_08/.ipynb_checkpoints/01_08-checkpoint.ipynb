```python
# merge:빠름, sql명령문과 비슷
help(pd.merge)
```

```python
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
# dfleft, dfright를 key변수를 기준으로 merge
# default가 inner
pd.merge(dfleft,dfright, how="left", on="KEY")
```

```python
pd.merge(dfleft,dfright,how="left") # default가 inner
# left => 왼쪽에 있는 DataFrame은 다 나오고 오른쪽은 공통된 영역만 출력
pd.merge(dfleft,dfright,how="right")
```

```python
pd.merge(dfleft,dfright, how="left", on="A") 
# on 다음에 오는것은 둘다 갖고 있는 것이여야 한다.
# 그래서 에러 뜸 이건
```

```python
pd.merge(dfleft,dfright, how="right", on="KEY")
```

```python
pd.merge(dfleft,dfright, how="outer", on="KEY", indicator=True)
```

```python
pd.merge(dfleft,dfright, how="outer", on="KEY", indicator="indicator_info")
```

```python
# index와 key를 조합하여 merge
dfleft = DataFrame({"KEY":['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},index=['K2', 'K3', 'K4', 'K5'])
```

```python
height=[0,0,3,5,2,4,9,0,6,4,0,6,0,0]
num=[-2,-1,1,2] # 2이상의 공간이 확보 되어있는 판단하기 위한 리스트
zero=np.zeros((15,15)) # 15x15 영행렬 생성
cnt=0
for i in height:
    if i !=0:
        zero[:14-i:-1,cnt]=1 # 영행렬에서 리스트 만큼의 높이의 빌딩 층수를 1로 찍어 줌 // 빌딩의 층 = 1
    cnt+=1
print(zero,"\n\n")    

dump=0
cnt=0

for i in range(15):
    for j in range(15):
        if zero[i][j]==1: # 행렬에서 1을 만나면
            cnt=0
            for k in num:
                if zero[i][j-k]==0:  # 1의 좌우 2칸이 전부 0인지 판단
                    cnt+=1
                if cnt==4:
                    dump+=1
# print(cnt)
print(dump)    
```
