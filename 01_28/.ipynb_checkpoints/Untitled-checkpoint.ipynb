```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ax=plt.subplots()
tips=sns.load_dataset("tips")
ax=sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title("Kernel Density Plot")
ax.set_xlabel("TB")
ax.set_ylabel("Tip")
```

```python
joint=sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel="TB", ylabel="Tip")
```

```python
#   
kde, ax=plt.subplots()
sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True)
ax.set_title("Kernel Density Plot")
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
#digits
samples=[0,10,20,30,1,11,21,31]
```

```python
d=[]
for i in range(8):
    d.append(digits.images[samples[i]])
```

```python
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) # i+1 =>   ?
    plt.imshow(d[i], cmap=plt.cm.bone_r)
    plt.title("image{}".format(i+1))
    plt.grid(False)
    plt.xticks([]);plt.yticks([])
```

```python
# 2  => 64  => 1 
v=[]
for i in range(8):
    v.append(d[i].reshape(64,1))
    
plt.figure(figsize=(16,10))
for i in range(8):
    plt.subplot(1,8,i+1) # i+1 =>   ?
    plt.imshow(v[i], cmap=plt.cm.bone_r)
    plt.title("vector{}".format(i+1))
    plt.grid(False)
    plt.xticks([]);plt.yticks([])
```

```python
digits=load_digits()
d1=digits.images[0]
d2=digits.images[10]
d3=digits.images[1]
d4=digits.images[11]

v1=d1.reshape(64,1)
v2=d2.reshape(64,1)
v3=d3.reshape(64,1)
v4=d4.reshape(64,1)
```

```python
plt.figure(figsize=(9,9)) #  1 1
import matplotlib.gridspec as gridspec

```

```python
gs=gridspec.GridSpec(1,8,height_ratios=[1],width_ratios=[9,1,9,1,9,1,9,1])
```

```python
for i in range(4):
    plt.subplot(gs[2*i])
    plt.imshow(eval("d"+str(i+1)), cmap=plt.cm.bone_r)
    plt.subplot(gs[2*i+1])
    plt.imshow(eval("v"+str(i+1)), cmap=plt.cm.bone_r)
```

```python
v1
(v1.T@v2)[0][0] # 3064 v1, v2  
(v1.T@v3)[0][0] # 1866 v1, v3  
(v1.T@v4)[0][0] # 1883 v1, v4  
```
