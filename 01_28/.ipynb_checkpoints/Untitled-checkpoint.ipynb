```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ax=plt.subplots()
tips=sns.load_dataset("tips")
ax=sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title("Kernel Density Plot")
ax.set_xlabel("TB")
ax.set_ylabel("Tip")
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
#digits
samples=[0,10,20,30,1,11,21,31]
```

```python
d=[]
for i in range(8):
    d.append(digits.images[samples[i]])
```

```python
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) # i+1 => 몇 번째 칸인지?
    plt.imshow(d[i], cmap=plt.cm.bone_r)
    plt.title("image{}".format(i+1))
    plt.grid(False)
    plt.xticks([]);plt.yticks([])
```

```python
# 2차원 이미지 => 64차원 데이터 => 1차원 벡터
v=[]
for i in range(8):
    v.append(d[i].reshape(64,1))
    
plt.figure(figsize=(16,10))
for i in range(8):
    plt.subplot(1,8,i+1) # i+1 => 몇 번째 칸인지?
    plt.imshow(v[i], cmap=plt.cm.bone_r)
    plt.title("vector{}".format(i+1))
    plt.grid(False)
    plt.xticks([]);plt.yticks([])
```

```python
# 스칼라와 벡터/행렬간 곱셈

# 선형조합 : 벡터/행렬에 스칼라값을 곱한 후 더하거나 뺀 것을 벡터/행렬의 선형조합이라고 함

# 내적 : 두 벡터의 곱셈
# 내적조건- 1. 두 벡터의 차원이 같아야 한다. // 2. 행 벡터 = 열 벡터 
```

```python
x=np.array([1,2,3]).reshape(3,1)
y=np.array([4,5,6]).reshape(3,1)
# np.dot(x.T,y)
x.T@y # @ => 내적
```

```python
# 데이터 벡터, 가중치벡터 => 데이터 벡터의 가중치 합
```

```python
"""
벡터의 내적은 두 벡터의 유사도를 계산하는데 활용이 됨
유사도? 두 벡터가 닮은 정도
"""
```

```python
digits=load_digits()
d1=digits.images[0]
d2=digits.images[10]
d3=digits.images[1]
d4=digits.images[11]

v1=d1.reshape(64,1)
v2=d2.reshape(64,1)
v3=d3.reshape(64,1)
v4=d4.reshape(64,1)
```

```python
plt.figure(figsize=(9,9)) # 숫자 1당 1픽셀
import matplotlib.gridspec as gridspec

```

```python
gs=gridspec.GridSpec(1,8,height_ratios=[1],width_ratios=[9,1,9,1,9,1,9,1])
```

```python
for i in range(4):
    plt.subplot(gs[2*i])
    plt.imshow(eval("d"+str(i+1)), cmap=plt.cm.bone_r)
    plt.subplot(gs[2*i+1])
    plt.imshow(eval("v"+str(i+1)), cmap=plt.cm.bone_r)
```

```python
v1
(v1.T@v2)[0][0] # 3064 v1, v2 벡터간 내적
(v1.T@v3)[0][0] # 1866 v1, v3 벡터간 내적
(v1.T@v4)[0][0] # 1883 v1, v4 벡터간 내적
```
