```python
height=[5,8,3,1,5,6,9,9,2,2,4]

def dump():
    min_num=height[0]
    max_num=height[0]
    chn_min=0
    chn_max=0
    for i in height:
        if i <= min_num:
            min_num=i
        if i >= max_num:
            max_num=i
    chn_min=height.index(min_num)
    chn_max=height.index(max_num)
    height[chn_min]=height[chn_min]+1
    height[chn_max]=height[chn_max]-1
    print(height,"최고점-최저점=>",max(height),"-",min(height),"=",max(height)-min(height))

for i in range(10):
    dump()
```
